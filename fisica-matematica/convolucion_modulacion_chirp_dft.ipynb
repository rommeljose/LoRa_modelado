### Convolución en la Modulación Chirp y la DFT

En el procesamiento de señales, la **convolución** aparece en varios aspectos. En este cuaderno explicaremos 
dónde y cómo se utiliza la convolución en el contexto de la modulación chirp y la Transformada Discreta de Fourier (DFT).

#### Relación entre la Convolución y la DFT

En la teoría de Fourier, existe una relación clave llamada el **teorema de la convolución**, que establece lo siguiente:

1. **Convolución en el dominio del tiempo** es equivalente a **multiplicación en el dominio de la frecuencia**.
2. **Multiplicación en el dominio del tiempo** es equivalente a **convolución en el dominio de la frecuencia**.

Esta relación es fundamental para comprender cómo las señales moduladas, como las usadas en sistemas LoRa (chirp),
pueden ser procesadas en el dominio del tiempo y la frecuencia.

#### 1. Convolución implícita en el proceso de "dechirping"

En la modulación chirp, el proceso de **dechirping** implica multiplicar la señal recibida por un chirp de referencia
invertido en frecuencia. Esta multiplicación en el tiempo tiene un efecto equivalente a una **convolución** en el dominio
de la frecuencia.

- La señal recibida, `r(t)`, contiene componentes de frecuencia que varían con el tiempo.
- El chirp de referencia, utilizado para el dechirping, actúa como una convolución inversa, alineando las frecuencias de
la señal recibida.
- Esto facilita la identificación de los símbolos transmitidos una vez aplicada la Transformada Discreta de Fourier (DFT).

#### 2. Convolución en la detección de símbolos

Después del proceso de dechirping, aplicamos una **DFT** para identificar el símbolo transmitido.
Este paso puede verse como una operación de **correlación** (que es similar a una convolución) entre la señal recibida y
los perfiles de frecuencia esperados para cada símbolo.

La DFT está definida por:

$$
X[k] = \sum_{n=0}^{N-1} r[n] \cdot e^{-j2\pi \frac{kn}{N}}
$$

La DFT transforma la señal al dominio de la frecuencia, permitiendo identificar los símbolos transmitidos mediante 
la correlación con las posibles frecuencias.

### Resumen de Convolución en la Modulación Chirp y la DFT

- **Convolución en el tiempo**: Aparece implícitamente en el proceso de dechirping. Multiplicamos la señal recibida por 
un chirp de referencia invertido, lo que resulta en una alineación de las frecuencias.
  
- **Convolución en el dominio de la frecuencia**: La DFT actúa como una operación que identifica la frecuencia dominante
del símbolo transmitido, equivalente a una convolución con perfiles de frecuencia conocidos.
